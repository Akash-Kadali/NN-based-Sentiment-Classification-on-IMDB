```python
# 1) Mount Drive
from google.colab import drive
drive.mount('/content/drive')

PROJECT = "/content/drive/MyDrive/NLP3"

import os
os.environ["PROJECT"] = PROJECT

# 2) Folders + package init
os.makedirs(f"{PROJECT}/results/plots", exist_ok=True)
os.makedirs(f"{PROJECT}/results/experiments", exist_ok=True)
os.makedirs(f"{PROJECT}/logs", exist_ok=True)
open(f"{PROJECT}/__init__.py", "a").close()

# 3) Install deps
!pip -q install torch torchvision torchaudio pandas numpy scikit-learn matplotlib tqdm nltk psutil

# 4) NLTK data (including punkt_tab for newer NLTK)
import nltk
os.environ["NLTK_DATA"] = "/content/drive/MyDrive/nltk_data"
nltk.data.path.append("/content/drive/MyDrive/nltk_data")
nltk.download('punkt', download_dir=os.environ["NLTK_DATA"])
nltk.download('punkt_tab', download_dir=os.environ["NLTK_DATA"])
nltk.download('stopwords', download_dir=os.environ["NLTK_DATA"])

# 5) Project on PYTHONPATH
%cd /content
%env PYTHONPATH=/content/drive/MyDrive

# 6) Define 20 experiments (config list)
configs = []

# Group A: seq_len sweep for LSTM + Adam (3 runs)
for L in [50, 100, 200]:
    configs.append(dict(
        group="A_seq_len_LSTM",
        arch="lstm",
        seq_len=L,
        optimizer="adam",
        activation="tanh",
        grad_clip=0.0,
        seed=42,
        hidden_size=128,
    ))

# Group B: arch sweep at seq_len=200, Adam (3 runs)
for arch in ["rnn", "lstm", "bilstm"]:
    configs.append(dict(
        group="B_arch",
        arch=arch,
        seq_len=200,
        optimizer="adam",
        activation="tanh",
        grad_clip=0.0,
        seed=42,
        hidden_size=128,
    ))

# Group C: optimizer sweep for BiLSTM at L=200 (4 runs)
for opt in ["adam", "adamw", "sgd", "rmsprop"]:
    configs.append(dict(
        group="C_opt_bilstm",
        arch="bilstm",
        seq_len=200,
        optimizer=opt,
        activation="tanh",
        grad_clip=0.0,
        seed=42,
        hidden_size=128,
    ))

# Group D: activation sweep for BiLSTM + AdamW at L=200 (3 runs)
for act in ["tanh", "relu", "sigmoid"]:
    configs.append(dict(
        group="D_act_bilstm_adamw",
        arch="bilstm",
        seq_len=200,
        optimizer="adamw",
        activation=act,
        grad_clip=0.0,
        seed=42,
        hidden_size=128,
    ))

# Group E: grad_clip sweep for BiLSTM + AdamW + tanh at L=200 (2 runs)
for gc in [0.0, 1.0]:
    configs.append(dict(
        group="E_gradclip",
        arch="bilstm",
        seq_len=200,
        optimizer="adamw",
        activation="tanh",
        grad_clip=gc,
        seed=42,
        hidden_size=128,
    ))

# Group F: seed sweep for best-ish config (BiLSTM + AdamW + tanh + gc=1) (3 runs)
for sd in [0, 1, 2]:
    configs.append(dict(
        group="F_seed",
        arch="bilstm",
        seq_len=200,
        optimizer="adamw",
        activation="tanh",
        grad_clip=1.0,
        seed=sd,
        hidden_size=128,
    ))

# Group G: capacity variation for BiLSTM + AdamW + tanh + gc=1 (2 runs)
for hs in [64, 256]:
    configs.append(dict(
        group="G_capacity",
        arch="bilstm",
        seq_len=200,
        optimizer="adamw",
        activation="tanh",
        grad_clip=1.0,
        seed=42,
        hidden_size=hs,
    ))

print("Total configs:", len(configs))  # should be 20

# 7) Run all experiments
for i, cfg in enumerate(configs, 1):
    print(
        f"\n=== RUN {i}/{len(configs)} | group={cfg['group']} | "
        f"arch={cfg['arch']} | L={cfg['seq_len']} | opt={cfg['optimizer']} | "
        f"act={cfg['activation']} | gc={cfg['grad_clip']} | seed={cfg['seed']} | "
        f"hidden={cfg['hidden_size']} ==="
    )
    !python -m NLP3.train \
      --mode single \
      --arch {cfg['arch']} \
      --activation {cfg['activation']} \
      --optimizer {cfg['optimizer']} \
      --seq-len {cfg['seq_len']} \
      --grad-clip {cfg['grad_clip']} \
      --epochs 5 \
      --batch-size 512 \
      --num-workers 4 \
      --prefetch-factor 4 \
      --persistent-workers 1 \
      --hidden-size {cfg['hidden_size']} \
      --emb-dim 128 \
      --num-layers 2 \
      --dropout 0.3 \
      --lr 2e-3 \
      --weight-decay 0.01 \
      --seed {cfg['seed']} \
      --results-dir "$PROJECT/results"

# 8) Generate required plots (NOTE: evaluate.py does NOT take --results-dir)
!python -m NLP3.evaluate \
  --plot acc_f1_vs_seq_len \
  --arch lstm --activation tanh --optimizer adam --grad-clip 0 \
  --metrics "$PROJECT/results/metrics.csv" \
  --experiments-dir "$PROJECT/results/experiments" \
  --out-dir "$PROJECT/results/plots"

!python -m NLP3.evaluate \
  --plot best_worst_losses \
  --metrics "$PROJECT/results/metrics.csv" \
  --experiments-dir "$PROJECT/results/experiments" \
  --out-dir "$PROJECT/results/plots"

# 9) Inspect results and confirm images exist
import pandas as pd
from IPython.display import display, Image
from pathlib import Path

metrics_path = f"{PROJECT}/results/metrics.csv"
print("\nLooking for metrics at:", metrics_path)
df = pd.read_csv(metrics_path)
display(df.sort_values("F1", ascending=False).head(20))

p1 = Path(f"{PROJECT}/results/plots/acc_f1_vs_seq_len.png")
print("acc_f1_vs_seq_len.png exists:", p1.exists())
if p1.exists():
    display(Image(str(p1)))

for name in ["loss_curves_best.png", "loss_curves_worst.png"]:
    p = Path(f"{PROJECT}/results/plots/{name}")
    print(f"{name} exists:", p.exists())
    if p.exists():
        display(Image(str(p)))
```
